```python
df = df.drop('id',axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
```

```python
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
import math
df['price_log'] = df['price'].apply(lambda x: math.log(x)) 
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(x='month',y='price_log',data=df)
```

```python
df = df.drop('date',axis=1)
```

```python
df.head()
```

```python
# df['zipcode'].value_counts()
df = df.drop('zipcode',axis=1)
```

```python
df['bed*bath'] = df['bedrooms']*df['bathrooms']
```

```python
old_df = df
old_df = old_df.drop('bed*bath',axis=1)
old_df = old_df.drop('price_log',axis=1)
```

```python
old_df.head()
df = df.drop('price',axis=1)
```

```python
X = df.drop('price_log',axis=1).values # .values returns the np array under the df.
y = df['price_log'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test) # don't fit to the test set
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape # 19 features means our NN layer should have 19 neurons.
```

```python
model = Sequential()

model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
# Could be overkill to do these 4 layers

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=128,epochs=500)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot() # So we could have used more epochs - but this is fine
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
# So on average we are off by 0.136 (ln difference)
# For a house that costs $1,000,000, this difference would be off by $145,000, but the diff would only be $14,000 for a $100k prop
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
```
